```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
f1 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Temp_Increase.xlsx') 
f2 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Precip_Increase.xlsx') 
f3 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Max_Temp_Anomaly.xlsx') 
f4 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Max_Rain_Fall.xlsx') 
f5 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Max_Drought_Index.xlsx') 
f6 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Heatwave.xlsx') 
f7 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/CostLength.xlsx') 
f8 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Sealevel.xlsx')
f9 = pd.read_excel('/Users/ilyasozgur/Desktop/OutputforProject/Average Elevation.xlsx')
```

```python
f7.rename({'Coast length (km) (1)': 'Coast Length'}, axis=1, inplace=True)
f8.rename({'Sea Level Increase cm': 'Inc. Sea Level'}, axis=1, inplace=True)
```

```python
f1 = f1.merge(f2, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase']]

f1 = f1.merge(f3, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly']]

f1 = f1.merge(f4, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall']]

f1 = f1.merge(f5, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall','Max Drought']]

f1 = f1.merge(f6, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall','Max Drought','Heat Wave']]

f1 = f1.merge(f7, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall','Max Drought','Heat Wave','Coast Length']]

f1 = f1.merge(f8, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall','Max Drought','Heat Wave','Coast Length','Inc. Sea Level']]

f1 = f1.merge(f9, on='Country',how = 'left')
f1 = f1[['Country','Temp Increase','Precip Increase','Max Temp Anomaly','Max RainFall','Max Drought','Heat Wave','Coast Length','Inc. Sea Level','Average Elevation']]
```

```python
f1['Coast Length'] = f1['Coast Length'].fillna(value=0)
f1['Inc. Sea Level'] = f1['Inc. Sea Level'].fillna(value=0)
```

```python
f1['Coast Length'] = f1['Coast Length'].astype(float)
f1['Inc. Sea Level'] = f1['Inc. Sea Level'].astype(float)
f1['Average Elevation'] = f1['Average Elevation'].astype(float)
```

```python
f1['Temp. Inc. Rank1']       = f1['Temp Increase'].rank(method='min',ascending=False)
f1['Prec. Inc. Rank2']       = f1['Precip Increase'].rank(method='min',ascending=False)
f1['Max Temp. Inc. Rank3']   = f1['Max Temp Anomaly'].rank(method='min',ascending=False)
f1['Max RainFall Rank4']     = f1['Max RainFall'].rank(method='min',ascending=False)
f1['Max Drought Rank5']      = f1['Max Drought'].rank(method='min',ascending=False)
f1['Heat Wave Rank6']        = f1['Heat Wave'].rank(method='min',ascending=False)
f1['Coast Length Rank7']     = f1['Coast Length'].rank(method='min',ascending=False)
f1['Inc. Sea Level Rank8']   = f1['Inc. Sea Level'].rank(method='min',ascending=False)
f1['Average Elevation Rank9']      = f1['Average Elevation'].rank(method='min',ascending=True)

f1['Total Rank']   = f1[['Temp. Inc. Rank1','Prec. Inc. Rank2','Max Temp. Inc. Rank3','Max RainFall Rank4',
                         'Max Drought Rank5','Heat Wave Rank6','Coast Length Rank7','Inc. Sea Level Rank8',
                         'Average Elevation Rank9']].mean(axis=1)


f1  = f1.sort_values(by=['Total Rank'], ascending=True).round(decimals=2)
```
